###### PARENTESI QUADRE

A questo punto dobbiamo prendere questa espressione e dobbiamo dirgli: filtra su tutto il dataframe, non solo sulla colonna. Per farlo si usano le parentesi quadre. Quindi sul dataframe "props" voglio che prendi la colonna "prop.name" (da aggiungere prima di essa sempre props.) e mi restituisci solamente i valori uguali a "Proposition 060 blabla".
Chiamerò questo nuovo elemento "prop".

```python
prop = props[props.prop_name == "PROPOSITION 060- ADULT FILMS. CONDOMS. HEALTH REQUIREMENTS. INITIATIVE STATUTE."]
```

```python
prop.info()
```

```python
contribs.info()
```

##### UNIONI

Prima di fare un cosiddetto <i>merge</i> abbiamo visto, usando il metodo info(), come si chiamavano le colonne e abbiamo visto che in comune c'era "calaccess_committee_id"

Chiamerò "merged" la fusione tra i database. Quindi scriverò pd.merge(NOME DEL PRIMO DATABASE detto anche <i>argument</i>, NOME DEL SECONDO DATABASE, anch'esso detto <i>argument</i>). Lo faccio tra parentesi tonde, separando i nomi dei database con virgola semplice, e poi scriverò <b>on</b> per dargli il nome dell'argument che lega i due database. 

```python
merged = pd.merge(prop, contribs, on="calaccess_committee_id")
```

```python
merged.head()
```

```python
merged.info()
```

##### SOMMA TOTALE

<b>sum</b> e’ un metodo, e come tale deve essere eseguito con parentesi aperta e chiusa

A questo punto voglio vedere quanti supportano e quanti non supportano la misura, quindi faccio un bel .value_counts()

```python
merged.committee_position.value_counts()
```

Chiamerò <i>support</i> i valori della colonna "committee_position" che sono uguali a "support", e lo stesso faccio con "oppose"

```python
support = merged[merged.committee_position == "SUPPORT"]
```

```python
oppose = merged[merged.committee_position == "OPPOSE"]
```

```python
support.amount.sum()
```

```python
oppose.amount.sum()
```

##### CHI SONO I LARGEST CONTRIBUTORS 

Nell'oggetto "merged" ordiniamo i valori - ovvero sort_values() - con il parametro "amount"
Il filtro iniziale va dal più piccolo al più grande (ASCENDING), mentre noi vogliamo ordine DESCENDING, quindi cambiamo la linea di codice in:

merged.sort_values(“amount”, ascending=False)

<b>False con la F maiuscola per operatore booleano</b>

oppose.sort_values("amount", ascending=False)

Per avere la top 5, aggiungi .head()

```python
merged.sort_values("amount", ascending=False)
```

```python
oppose.sort_values("amount", ascending=False).head()
```

##### GROUP BY
Per vedere la somma delle contribuzioni, utile quando hai un sacco di valori che vuoi raggruppare. Vogliamo scoprire chi ha dato piu’ soldi e meno alla prop 60 per i condom obbligatori nei film porno. 

<i>merged.groupby(“committee_name_x”)</i>


non succede nulla ma se faccio: 

<i>merged.groupby("committee_name_x").amount.sum()</i>


Raggruppa e pulisce tutti i nomi di committee e li mette in lista disordinata, quindi


<i>merged.groupby("committee_name_x").amount.sum().reset_index()</i>


a questo punto li mette in lista disordinata, devo aggiungere 


<i>merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)</i>


Per capire chi sono i nomi e cognomi, ovvero <b>per unire due distinti valori</b>, faccio cosi:

<i>merged.groupby(["contributor_firstname","contributor_lastname"]).amount.sum().reset_index().sort_values("amount", ascending=False)</i>

Si usa la <b>parentesi quadra dentro la tonda</b> per raggruppare una lista. Infine, per capire se sono pro o contro, basta aggiungere: 

merged.groupby(["contributor_firstname","contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)

```python
merged.groupby("committee_name_x").amount.sum().reset_index()
```

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
merged.groupby(["contributor_firstname","contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```
